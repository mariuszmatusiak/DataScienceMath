## Natural logarithm approximation, Compound interest

```python
from math import exp
from sympy import *

# Obliczanie odsetek ciągłych 
# Dziennnie, przez dwa lata
P = 100 # Kapitał, kwota początkowa
r = 0.2 # roczna stopa procentowa
n = 365 # liczba okresów do podziału (dni)
t = 2 # przedział czasu (liczba lat)

A = P * ((1 + r/n) ** (n * t))
print(A)

# Co minutę
n = n * 24 * 60
A = P * ((1 + r/n) ** (n * t))
print(A)

# Cont.
A = P * exp(r*t)
print(A)

# Symbolic
n = symbols('n')
f = ((1+1/n)**n)
f_exp = limit(f, n, oo)
print(f_exp)
print(f_exp.evalf())
```

## Derivatives, integrals approximations

https://www.desmos.com/calculator/mccllupiak?lang=pl

```python
def derivative_approx(fun, x, step = 0.001):
    result = (fun(x + step) - fun(x)) / ((x + step) - x)
    return result

def test_fun(x):
    return ((x ** 2) - 2)

print (derivative_approx(test_fun, 2))
```

### Derivative in SymPy

```python
x = symbols('x')
f = x**2 - 2
dfdx = diff(f)
print(dfdx)
# Check value in point x0
x0 = 2
print(dfdx.subs(x, x0))
```

### Partial derivative

```python
from sympy import *
from sympy.plotting import plot3d
x, y = symbols('x y')
fun = 2 * x ** 3 + 3 * y ** 3
dfdx = diff(fun, x)
dfdy = diff(fun, y)
print("df/dx = " + str(dfdx))
print("df/dy = " + str(dfdy))
plot3d(fun)

```
